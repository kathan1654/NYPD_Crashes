```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Data= pd.read_csv('https://data.cityofnewyork.us/api/views/h9gi-nx95/rows.csv?accessType=DOWNLOAD&bom=true&format=true')
```

```python
Data.info()
```

```python
# Column header titles cleaning/renaming
Data.rename(columns = { 'CRASH DATE'
    
                        'ZIP CODE'          : 'ZIP_CODE',
                       'ON STREET NAME'    : 'STREET_ON',
                       'CROSS STREET NAME' : 'STREET_CROSS',
                       'OFF STREET NAME'   : 'STREET_OFF',
                       'NUMBER OF PERSONS INJURED'     : 'TOTAL_INJURED',
                       'NUMBER OF PERSONS KILLED'      : 'TOTAL_KILLED',
                       'NUMBER OF PEDESTRIANS INJURED' : 'PED_INJURED',
                       'NUMBER OF PEDESTRIANS KILLED'  : 'PED_KILLED',
                       'NUMBER OF CYCLIST INJURED'     : 'CYC_INJURED',
                       'NUMBER OF CYCLIST KILLED'      : 'CYC_KILLED',
                       'NUMBER OF MOTORIST INJURED'    : 'MOTO_INJURED',
                       'NUMBER OF MOTORIST KILLED'     : 'MOTO_KILLED',
                       'CONTRIBUTING FACTOR VEHICLE 1' : 'VEH_FACTOR_1',
                       'CONTRIBUTING FACTOR VEHICLE 2' : 'VEH_FACTOR_2',
                       'CONTRIBUTING FACTOR VEHICLE 3' : 'VEH_FACTOR_3',
                       'CONTRIBUTING FACTOR VEHICLE 4' : 'VEH_FACTOR_4',
                       'CONTRIBUTING FACTOR VEHICLE 5' : 'VEH_FACTOR_5',
                       'UNIQUE KEY' : 'UNIQUE_KEY',
                       'VEHICLE TYPE CODE 1' : 'VEH_TYPE_1',
                       'VEHICLE TYPE CODE 2' : 'VEH_TYPE_2',
                       'VEHICLE TYPE CODE 3' : 'VEH_TYPE_3',
                       'VEHICLE TYPE CODE 4' : 'VEH_TYPE_4',
                       'VEHICLE TYPE CODE 5' : 'VEH_TYPE_5'},
           inplace = True) 
# Missing values in columns
Data.isna().sum()
```

```python
Data.head()
```

```python
# Fill all blank values in column Borough as around 1/3 data is missing
# If a value is NaN it will be NYC
Data.loc[Data['BOROUGH'].isnull(), 'BOROUGH'] = 'NYC'
```

```python
Data.isna().sum()
```

```python
Data = Data.dropna(axis=0, subset=['TOTAL_INJURED'])
```

```python
Data.isna().sum()
```

```python
Data = Data.dropna(axis=0, subset=['TOTAL_KILLED'])
```

```python
# Combine DATE and TIME column to transform Series to DateTime needed for further analysis
Data['DATE'] = Data['CRASH DATE'] + ' ' + Data['CRASH TIME']
```

```python
Data.info()
```

```python
# Convert string to DateTime
Data['DATE'] = pd.to_datetime(Data.DATE)
```

```python
Data.info()
```

```python
Data.head()
```

```python
Data['Hour']=Data['DATE'].dt.hour
```

```python
Data['Hour']
```

```python
Interval = []
for row in Data['Hour']:
        if row < 6:   Interval.append('Night')
        elif row < 12:  Interval.append('Morning')
        elif row <18:  Interval.append('Noon')
            
        else:           Interval.append('Evening')
Data['Interval'] = Interval
Data.head() 
```

```python
Data.info()
```

```python
Data.head()
```

```python
Data.drop(['CRASH DATE','CRASH TIME','ZIP CODE','LATITUDE','LONGITUDE','LOCATION','STREET_ON','STREET_CROSS','STREET_OFF','VEH_FACTOR_1','VEH_FACTOR_2','VEH_FACTOR_3','VEH_FACTOR_4','VEH_FACTOR_5','COLLISION_ID','Hour'],inplace=True,axis=1)
```

```python
Data.head()
```

```python
Data.info()
```

```python
Data['TOTAL_INVOLVED']= Data['TOTAL_INJURED'] + Data['TOTAL_KILLED']
```

```python
Data.describe()
```

```python
Risk = []
for row in Data['TOTAL_INVOLVED']:
        if row > 0:  Risk.append('Low')
        elif row > 10:  Risk.append('Medium')
        elif row > 20:  Risk.append('High')
            
        else:            Risk.append('None')
Data['Risk'] = Risk
  
```

```python
Data.head()
```

```python
Data.drop(['VEH_TYPE_1','VEH_TYPE_2','VEH_TYPE_3','VEH_TYPE_4','VEH_TYPE_5'],inplace=True,axis=1)
```

```python
Data.info()
```

```python
Data.drop(['TOTAL_INVOLVED'],inplace=True,axis=1)
```

```python
TOTAL_INJURED = []
for row in Data['TOTAL_INJURED']:
        if row ==0:   TOTAL_INJURED.append('NO')
            
        else:           TOTAL_INJURED.append('YES')
Data['TOTAL_INJURED'] = TOTAL_INJURED     
```

```python
TOTAL_KILLED = []
for row in Data['TOTAL_KILLED']:
        if row ==0:   TOTAL_KILLED.append('NO')
            
        else:           TOTAL_KILLED.append('YES')
Data['TOTAL_KILLED'] = TOTAL_KILLED 
```

```python
PED_KILLED = []
for row in Data['PED_KILLED']:
        if row ==0:  PED_KILLED.append('NO')
            
        else:           PED_KILLED.append('YES')
Data['PED_KILLED'] = PED_KILLED 
```

```python
PED_INJURED = []
for row in Data['PED_INJURED']:
        if row ==0:   PED_INJURED.append('NO')
            
        else:            PED_INJURED.append('YES')
Data['PED_INJURED'] =  PED_INJURED
```

```python
CYC_INJURED = []
for row in Data['CYC_INJURED']:
        if row ==0:   CYC_INJURED.append('NO')
            
        else:            CYC_INJURED.append('YES')
Data['CYC_INJURED'] =  CYC_INJURED
```

```python
CYC_KILLED = []
for row in Data['CYC_KILLED']:
        if row ==0:  CYC_KILLED.append('NO')
            
        else:           CYC_KILLED.append('YES')
Data['CYC_KILLED'] =  CYC_KILLED
```

```python
MOTO_INJURED = []
for row in Data['MOTO_INJURED']:
        if row ==0:  MOTO_INJURED.append('NO')
            
        else:          MOTO_INJURED.append('YES')
Data['MOTO_INJURED'] =  MOTO_INJURED
```

```python
MOTO_KILLED = []
for row in Data['MOTO_KILLED']:
        if row ==0:  MOTO_KILLED .append('NO')
            
        else:          MOTO_KILLED .append('YES')
Data['MOTO_KILLED'] =  MOTO_KILLED 
```
